```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,21)
```

```python
x
```

```python
y=2*x
```

```python
y
```

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
